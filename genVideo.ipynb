```python
####
# Filename:        genVideo.ipynb
# Date:            Jan 02 2017
# Last Edited by:  Gengshan Yang
# Description:     Generate videos of detection, tracking and pose estimation
#                  results. 
####

import os
import subprocess

def mvRes(imgPath, imgList, subDirName):
    """ move specified image results to given subfolder """
    dirName = os.path.join(imgPath, subDirName) 
    if os.path.exists(dirName):
        return
    else:
        os.makedirs(dirName)
        
    imgList = sorted(imgList)
    for it, name  in enumerate(imgList):
        a = os.path.join(imgPath, name)
        b = os.path.join(dirName, str(it).zfill(5) + ".jpg")
        os.rename(a, b)
        
def runFFMPEG(tbFilled):
    """ concat specified frames to a video, and clean subfloder """
    if os.path.exists('data/' + tbFilled + '.mp4'):
        return
    
    cmd = 'ffmpeg -f image2 -r 5 -i data/' + tbFilled + '/%05d.jpg -vcodec mpeg4 -y data/' + tbFilled + '.mp4'
    a = subprocess.check_output(cmd, shell=True, stderr=subprocess.STDOUT)
    print a
    
    """ remove image folder """
    cmd = 'rm -rf data/' + tbFilled
    a = subprocess.check_output(cmd, shell=True, stderr=subprocess.STDOUT)
    print a
```

```python
imgPath = "/data/gengshan/handProj/"  # specify output frame path
```

```python
""" get a list of image names """
imgList = os.listdir(imgPath)
detList = [x for x in imgList if ("det" in x) ]
trkList = [x for x in imgList if ("trk" in x) ]
posList = [x for x in imgList if ("pos" in x) ]
objPosList = {}
objIDs = list(set([x.split('_')[1] for x in posList]))
for ID in objIDs:
    objPosList[ID] = [x for x in posList if x.split('_')[1] == ID]
```

```python
mvRes(imgPath, detList, 'det')
mvRes(imgPath, trkList, 'trk')
for ID, val in objPosList.items():
    mvRes(imgPath, val, 'pos_' + ID)
```

```python
runFFMPEG('det')
runFFMPEG('trk')
for ID, val in objPosList.items():
    runFFMPEG('pos_' + ID)
```
